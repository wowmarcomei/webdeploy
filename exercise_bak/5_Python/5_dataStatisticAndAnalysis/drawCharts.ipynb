```python
import pymongo
import charts
```

```python
client = pymongo.MongoClient('localhost',27017)
ganjidb = client['ganjidb']
ganji_url_list = ganjidb['ganji_url_list']
ganji_item_info = ganjidb['ganji_item_info']
```

```python
item_list = []
for i in ganji_item_info.find():
#    print(i['title'],i['area'],i['want_person'],i['price'])
    item_list.append(i['area'].split('-')[-1])
area_index = list(set(item_list)) #去重,获得地区列表
print(area_index)
```

```python
post_times = []
for index in area_index:
    post_times.append(item_list.count(index))
print(post_times)
```

```python
def data_gen(types):
    length = 0
    if length <= len(area_index):
        for area,times in zip(area_index,post_times):
            data={
                'name':area,
                'data':[times],  #注意这里必须的键值对必须使用data作为key，value必须加括号[]
                'type':types
            }
            yield data
            length += 1
```

```python
for i in data_gen('column'):
    print(i)
```

```python
series = [data for data in data_gen('column')]
charts.plot(series,show='inline',options=dict(title=dict(text='Charts are awisome!!!')))
```
